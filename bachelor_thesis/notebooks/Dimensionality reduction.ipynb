```python
from sklearn.decomposition import NMF, TruncatedSVD
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
```

```python
X, y = fetch_openml('mnist_784', version=1, return_X_y=True)
X /= 255
```

```python
X = X.to_numpy()
```

```python
plt.gray()
fig, axes = plt.subplots(4, 4, figsize=(10, 10))
for i in range(4):
    for j in range(4):
        axes[i][j].matshow(X[4 * i + j].reshape(28, 28))
plt.tight_layout()
```

```python
nmf = NMF(n_components=64)
transformed = nmf.fit_transform(X)
transformed.shape
```

```python
plt.gray()
fig, axes = plt.subplots(8, 8, figsize=(24, 24))
for i in range(8):
    for j in range(8):
        axes[i][j].matshow(nmf.components_[8*i + j].reshape(28, 28))
plt.tight_layout()
```

```python
svd = TruncatedSVD(n_components=64)
svd_trand = svd.fit_transform(X)
svd.components_
```

```python
plt.gray()
fig, axes = plt.subplots(4, 4, figsize=(10, 10))
for i in range(4):
    for j in range(4):
        axes[i][j].matshow(transformed[4 * i + j].reshape(8, 8))
plt.tight_layout()
```

```python
nmfi = NMF(n_components=3)
nmfi.fit_transform(X[:100])
```
